```python
channel_df = pd.read_sql("SELECT * FROM public.\"Channel\"", conn)
video_df = pd.read_sql("SELECT * FROM public.\"Video\"", conn)
comment_df = pd.read_sql("SELECT * FROM public.\"Comments\"", conn)
```

비디오 테이블

좋아요 비율 : 비디오 좋아요수 / 비디오 조회수   
댓글 비율 : 댓글 수 / 비디오 조회수

```python
video_df['videoLikeCount'] = video_df['videoLikeCount'].astype(float)
video_df['videoViewCount'] = video_df['videoViewCount'].astype(float)
video_df['commentCount'] = video_df['commentCount'].astype(float)

video_df['like_ratio'] = video_df['videoLikeCount'] / video_df['videoViewCount']
video_df['comment_ratio'] = video_df['commentCount'] / video_df['videoViewCount']
```

채널 테이블

평균 조회수 : (채널 전체 조회수 / 비디오개수)   
구독자 대비 조회수 : (채널 전체 조회수 / 구독자 수)

```python
channel_df['subscriberCount'] = channel_df['subscriberCount'].astype(float)
channel_df['viewCount'] = channel_df['viewCount'].astype(float) 
channel_df['videoCount'] = channel_df['videoCount'].astype(float)

channel_df['sub_view_ratio'] = channel_df['viewCount'] / channel_df['subscriberCount']
channel_df['avg_views'] = channel_df['viewCount'] / channel_df['videoCount']
```

```python
result = video_df.groupby('channel_id').agg({
   'like_ratio': 'mean',
   'comment_ratio': 'mean'
}).join(
   channel_df.set_index('id')[['avg_views', 'sub_view_ratio']]
).round(4)

# avg_views를 실수 형태로 표시
pd.set_option('display.float_format', '{:.8f}'.format)

print(result)
```
