```python
import time
import requests
import json

class CompletionExecutor:
    def __init__(self, host, api_key, request_id):
        self._host = host
        self._api_key = api_key
        self._request_id = request_id
        self._max_retries = 5
        
    def execute(self, completion_request):
        headers = {
            "Authorization": self._api_key,
            "X-NCP-CLOVASTUDIO-REQUEST-ID": self._request_id,
            "Content-Type": "application/json; charset=utf-8",
        }

        retries = 0  # 현재 재시도 횟수
        while True:
            # POST 요청 보내기
            response = requests.post(
                self._host + "/testapp/v1/chat-completions/HCX-003",
                headers=headers,
                json=completion_request
            )

            # 응답 상태 확인
            if response.status_code == 200:
                response_data = response.json()
                print(response_data['result']["message"]["content"])
                return response_data['result']["message"]["content"]
            elif response.status_code == 429:  # Too Many Requests
                print(f"Rate limit exceeded. Retrying after {1} seconds...")
                time.sleep(10)
                retries+=1
    
# CompletionExecutor 객체 생성 및 실행
completion_executor = CompletionExecutor(
    host="https://clovastudio.stream.ntruss.com",
    api_key="Bearer nv-f5786fde571f424786ed0823986ca992h3P1",
    request_id="aca3cc2b98354bad93bbf15f4b63f616",
)
```

```python
# CLOVA Speech API 요청
def clova_speech_stt(file_path, invoke_url, secret_key):
    headers = {
        'X-CLOVASPEECH-API-KEY': secret_key
    }
    files = {
        'media': open(file_path+'.mp3', 'rb'),
        'params': (None, json.dumps({
            'language': 'ko-KR', #enko
            'completion': 'sync'
        }).encode('UTF-8'), 'application/json')
    }
    response = requests.post(f"{invoke_url}/recognizer/upload", headers=headers, files=files).json()['segments']
    text = []
    for segment in response:
        text.append(segment)
    text = ' '.join(text)
    with open(file_path+'.txt', 'w', encoding='utf-8-sig') as f:
        f.write(text)
    return text
```

```python
def get_text(text_path):
    with open(text_path, 'r', encoding='utf-8-sig') as f:
        text = f.read()
    return text

def text_add(prompt, text, genre, title):
    text_add = f"채널의 장르: {genre}\n제목: {title}\n스크립트: " + text
    formatted_prompt = [prompt[0], {"role": "user", "content": prompt[1]["content"].format(text_add)}]
    return formatted_prompt

def hcx_output(request_data):
    return completion_executor.execute(request_data)
```

```python
import pandas as pd
df = pd.read_csv("human_eval.csv")
df['base_발생가능성'], df['base_심각성'], df['base_output'] = None, None, None
df['oneshot_발생가능성'], df['oneshot_심각성'], df['oneshot_output'] = None, None, None
df['decompose_발생가능성'], df['decompose_심각성'], df['decompose_output'] = None, None, None
df.head()
```

```python
import re
def os_parsing(text):
    # 발생 가능성 추출
    occurrence_match = re.search(r"발생 가능성:\s*-\s*([가-힣]+)", text)
    if not occurrence_match:
        occurrence_match = re.search(r"발생 가능성:\s([가-힣]+)", text)
    occurrence_status = occurrence_match.group(1)  if occurrence_match else None # 낮음/중간/높음
    #occurrence_score = occurrence_match.group(2)   # 세부 점수
        
    severity_match = re.search(r"심각성:\s*-\s*([가-힣]+)", text)
    if not severity_match:
        severity_match = re.search(r"발생 가능성:\s([가-힣]+)", text)
    severity_status = severity_match.group(1)  if severity_match else None  # 낮음/중간/높음
    #severity_score = severity_match.group(2)   # 세부 점수
        
    print(occurrence_status, severity_status)
    return occurrence_status, severity_status
```

```python
import re
def parsing(text):
    # 발생 가능성의 추출
    occurrence_match = re.search(r"발생 가능성:\s*- 점수:\s*(\S+)", text)
    occurrence_score = occurrence_match.group(1) if occurrence_match else None
    
    # 심각성의 추출
    severity_match = re.search(r"심 각성:\s*- 점수:\s*(\S+)", text)
    severity_score = severity_match.group(1) if severity_match else None
    print(occurrence_score, severity_score)
    return occurrence_score, severity_score
```

```python
import time
print('시작점 : ', last_idx)
try:
    for i in df.index[last_idx:]:
        row = df.iloc[i]
        title = row['제목']
        genre = row['장르']
        script = row['스크립트 전체 내용']
        
        for type in ['base', 'oneshot', 'decompose']:
            print(type)
            if type=='base': prompt=base_PROMPT
            elif type=='oneshot': prompt=oneshot_PROMPT
            else: prompt=decompose_PROMPT
            
            text = text_add(prompt, script, genre, title)   # user prompt
            request_data = {
                "messages": text,
                "topP": 0.8,
                "topK": 0,
                "maxTokens": 1024,
                "temperature": 0.35,
                "repeatPenalty": 2.5,
                "stopBefore": [],
                "includeAiFilters": False,
                "seed": 104,
            }
            output = hcx_output(request_data)
            df.at[i, f'{type}_output'] = output   # output 먼저 저장
            if type == "decompose":
                occ, sev = parsing(output)
            else:
                occ, sev = os_parsing(output)
            if occ==None and sev==None:
                print(f'{title} 영상 나중에 너가 채워넣어라')
                continue
            
            df.at[i, f'{type}_발생가능성'], df.at[i, f'{type}_심각성'] = occ, sev
        time.sleep(5)
        last_idx = i
except Exception as e:
    print('Error:', {str(e)})
    print('last_idx : ', last_idx)
```
