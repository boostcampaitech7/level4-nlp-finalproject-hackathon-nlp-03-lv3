```python
import time
import requests
import json

class CompletionExecutor:
    def __init__(self, host, api_key, request_id):
        self._host = host
        self._api_key = api_key
        self._request_id = request_id
        self._max_retries = 5
        
    def execute(self, completion_request):
        headers = {
            "Authorization": self._api_key,
            "X-NCP-CLOVASTUDIO-REQUEST-ID": self._request_id,
            "Content-Type": "application/json; charset=utf-8",
        }

        retries = 0  # 현재 재시도 횟수
        while retries < self._max_retries:
            # POST 요청 보내기
            response = requests.post(
                self._host + "/testapp/v1/chat-completions/HCX-003",
                headers=headers,
                json=completion_request
            )

            # 응답 상태 확인
            if response.status_code == 200:
                response_data = response.json()
                return response_data['result']["message"]["content"]
            elif response.status_code == 429:  # Too Many Requests
                print(f"Rate limit exceeded. Retrying after {1} seconds...")
                time.sleep(10)
                #print(response.json())
                retries+=1
    
# CompletionExecutor 객체 생성 및 실행
completion_executor = CompletionExecutor(
    host="https://clovastudio.stream.ntruss.com",
    api_key="Bearer nv-f5786fde571f424786ed0823986ca992h3P1",
    request_id="aca3cc2b98354bad93bbf15f4b63f616",
)
```

```python
# CLOVA Speech API 요청
def clova_speech_stt(file_path, invoke_url, secret_key):
    headers = {
        'X-CLOVASPEECH-API-KEY': secret_key
    }
    files = {
        'media': open(file_path+'.mp3', 'rb'),
        'params': (None, json.dumps({
            'language': 'ko-KR', #enko
            'completion': 'sync'
        }).encode('UTF-8'), 'application/json')
    }
    response = requests.post(f"{invoke_url}/recognizer/upload", headers=headers, files=files).json()['segments']
    text = []
    for segment in response:
        text.append(segment)
    text = ' '.join(text)
    with open(file_path+'.txt', 'w', encoding='utf-8-sig') as f:
        f.write(text)
    return text
```

```python
def get_text(text_path):
    with open(text_path, 'r', encoding='utf-8-sig') as f:
        text = f.read()
    return text

def text_add(prompt, text, genre, title):
    text_add = f"채널의 장르: {genre}\n제목: {title}\n스크립트: " + text
    formatted_prompt = [prompt[0], {"role": "user", "content": prompt[1]["content"].format(text_add)}]
    return formatted_prompt

def hcx_output(request_data):
    return completion_executor.execute(request_data)
```

```python
import re
def parsing(text):
    # 발생 가능성 추출
    occurrence_match = re.search(r"발생 가능성:\s*-\s*([가-힣]+)", text)
    if not occurrence_match:
        occurrence_match = re.search(r"발생 가능성:\s([가-힣]+)", text)
        if not occurrence_match:
            print(text)
            occurrence_match = None
        else:
            occurrence_status = occurrence_match.group(1)  # 낮음/중간/높음
        #occurrence_score = occurrence_match.group(2)   # 세부 점수

    severity_match = re.search(r"심각성:\s*-\s*([가-힣]+)", text)
    if not severity_match:
        severity_match = re.search(r"발생 가능성:\s([가-힣]+)", text)
        if not severity_match:
            print(text)
            severity_match = None
        else:
            severity_status = severity_match.group(1)  # 낮음/중간/높음
            #severity_score = severity_match.group(2)   # 세부 점수
    print(occurrence_status, severity_status)
    return occurrence_status, severity_status
```
