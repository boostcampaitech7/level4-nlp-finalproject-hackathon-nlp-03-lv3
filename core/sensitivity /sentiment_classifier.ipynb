```python
import time
import requests

class CompletionExecutor:
    def __init__(self, host, api_key, request_id):
        self._host = host
        self._api_key = api_key
        self._request_id = request_id
        self._max_retries = 5

    def execute(self, completion_request):
        headers = {
            "Authorization": self._api_key,
            "X-NCP-CLOVASTUDIO-REQUEST-ID": self._request_id,
            "Content-Type": "application/json; charset=utf-8",
        }
        # POST 요청을 보냄 (스트리밍 대신 일반 요청)
        response = requests.post(
            self._host + "/testapp/v1/chat-completions/HCX-003",
            headers=headers,
            json=completion_request
        )

        # 응답 상태 확인
        if response.status_code == 200:
            response_data = response.json()
            # {'status': {'code': '20000', 'message': 'OK'}, 'result': {'message': {'role': 'assistant', 'content': '감성: 긍정'}, 'inputLength': 147, 'outputLength': 5, 'stopReason': 'stop_before', 'seed': 104}}
            return response_data['result']["message"]["content"]
        else:
            # 에러 처리
            raise Exception(f"Request failed with status code {response.status_code}: {response.text}")
        
    def execute_retry(self, completion_request):
        headers = {
            "Authorization": self._api_key,
            "X-NCP-CLOVASTUDIO-REQUEST-ID": self._request_id,
            "Content-Type": "application/json; charset=utf-8",
        }

        retries = 0  # 현재 재시도 횟수
        while retries < self._max_retries:
            # POST 요청 보내기
            response = requests.post(
                self._host + "/testapp/v1/chat-completions/HCX-003",
                headers=headers,
                json=completion_request
            )

            # 응답 상태 확인
            if response.status_code == 200:
                response_data = response.json()
                return response_data['result']["message"]["content"]
            elif response.status_code == 429:  # Too Many Requests
                print(f"Rate limit exceeded. Retrying after {1} seconds...")
                time.sleep(10)
                print(response.json())
                retries+=1
    
# CompletionExecutor 객체 생성 및 실행
completion_executor = CompletionExecutor(
    host="https://clovastudio.stream.ntruss.com",
    api_key="Bearer nv-f5786fde571f424786ed0823986ca992h3P1",
    request_id="aca3cc2b98354bad93bbf15f4b63f616",
)
```

```python
PROMPT = [
    {
        "role": "system",
        "content": """다음은 유튜브 댓글 입니다. 각 댓글의 감성을 분석하여 '긍정' 또는 '부정'으로 분류해 주세요.
        다음 지시사항을 참고해 주세요:
        1. 채널의 성격(예: 코미디, 여행, 엔터 등)에 따라 대중의 반응이 다를 수 있습니다. 따라서 채널의 성격을 고려하여 댓글을 평가하세요.
        2. 코미디 채널에서는 다소 부정적인 표현도 유머로 받아들여질 수 있으니 맥락을 고려하세요.

        결과는 다음 형식으로 출력해 주세요:
        감성: [긍정/부정]"""
    },
    {
        "role": "user",
        "content": """아래 데이터를 분석해 주세요:
        채널명: {}
        댓글 내용: {}"""
    }
]
```

```python

result = []
for _, group in df_v.groupby(['channelName']):
    print(_[0])
    vId_ = group.groupby(['vId'])['likeCount'].sum().index[:10].tolist()
    group_ = group[group['vId'].isin(vId_)]
    for _, row in group_.iterrows():
        print(_)
        channelName = row['channelName']
        if channelName == '싱긍벙글':
            channelName = '싱글벙글'
        vId = row['vId']
        cId = row['cId']
        likeCount = row['likeCount']
        commentPublishedAt = row['commentPublishedAt']
        
        textDisplay = row['textDisplay']
        textDisplay = textDisplay.split('a>')[-1].strip()
        textDisplay = str(textDisplay.replace('\n', ' '))
        textDisplay = str(textDisplay.replace('\n', ' '))
        formatted_prompt = [PROMPT[0], {"role": "user", "content": PROMPT[1]["content"].format(channelName, textDisplay)}]
        request_data = {
            "messages": formatted_prompt,
            "topP": 0.6,
            "topK": 0,
            "maxTokens": 16,
            "temperature": 0.10,
            "repeatPenalty": 1.2,
            "stopBefore": [],
            "includeAiFilters": False,
            "seed": 104,
        }
        pred = completion_executor.execute_retry(request_data)
        result.append([channelName, str(vId), str(cId), str(textDisplay), pred, likeCount, commentPublishedAt])
```
