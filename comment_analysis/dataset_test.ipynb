```python
import pandas as pd

kote = pd.read_csv('data/kote_relabeled.csv')
aihub = pd.read_csv('data/aihub_relabeled.csv')
```

```python
print(kote.head())
print(aihub.head())
```

```python
kote = kote[['text', 'final_label']]
kote = kote.rename(columns={'text': 'Text', 'final_label': 'Completion'})
kote
```

```python
kote = kote[['Text', 'Completion']]
kote.to_csv('kote.csv', index=False)
```

```python
aihub = aihub[['Sentence', 'Label']]
aihub = aihub.rename(columns={'Sentence': 'Text', 'Label': 'Completion'})
aihub
```

### kote + aihub

```python
df = pd.concat([kote, aihub], ignore_index=True)
df
```

```python
print(len(df[df['Completion']=='긍정']))
print(len(df[df['Completion']=='부정']))
print(len(df[df['Completion']=='중립']))
```

```python
# '긍정'과 '부정' 라벨을 25000개로 다운샘플링
df_positive_downsampled = df[df['Completion'] == '긍정'].sample(n=25000, random_state=42)
df_negative_downsampled = df[df['Completion'] == '부정'].sample(n=25000, random_state=42)
df_neutral = df[df['Completion'] == '중립']

# 다운샘플링된 데이터프레임들을 결합
downsampled_df = pd.concat([df_positive_downsampled, df_negative_downsampled, df_neutral])

downsampled_df = downsampled_df.sample(frac=1, random_state=42).reset_index(drop=True)
downsampled_df.to_csv('downsampled_three_emotions.csv', index=False)
```

```python
print(len(downsampled_df[downsampled_df['Completion']=='긍정']))
print(len(downsampled_df[downsampled_df['Completion']=='부정']))
print(len(downsampled_df[downsampled_df['Completion']=='중립']))
```

```python
# 샘플링된 인덱스 저장
positive_sampled_indices = df_positive_downsampled.index
negative_sampled_indices = df_negative_downsampled.index

# 남은 데이터에서 추가로 3500개 샘플링
remaining_positive = df[df['Completion'] == '긍정'].drop(positive_sampled_indices)
remaining_negative = df[df['Completion'] == '부정'].drop(negative_sampled_indices)

additional_positive = remaining_positive.sample(n=3500, random_state=42)
additional_negative = remaining_negative.sample(n=3500, random_state=42)

# 추가 샘플링된 데이터를 CSV로 저장
additional_samples = pd.concat([additional_positive, additional_negative])
additional_samples = additional_samples.sample(frac=1, random_state=42).reset_index(drop=True)
additional_samples.to_csv('test_three_emotions.csv', index=False)
```
